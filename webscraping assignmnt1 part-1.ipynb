# 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
page1=requests.get('http://www.imdb.com/chart/top')
```

```python
page1.content
```

```python
soup1=BeautifulSoup(page1.content,'html.parser')
```

```python
film_name=[]
imdb_rating=[]
yearofrelease=[]
```

```python
name=soup1.select('td.titleColumn')
name[0:100]
```

```python
rating=soup1.find_all('td',class_='ratingColumn imdbRating')
rating[0:100]
```

```python
year=soup1.find_all('span',class_='secondaryInfo')
year[0:100]
```

```python
for i in name:
    film_name.append(i.get_text().replace("\n",""))
    
film_name
```

```python
for i in rating:
    imdb_rating.append(i.get_text().replace("\n",""))
    
imdb_rating
```

```python
for i in year:
    yearofrelease.append(i.get_text().replace("\n",""))
    
yearofrelease
```

```python
import pandas as pd
IMDB=pd.DataFrame({})
IMDB['film']=film_name[0:100]
IMDB['Rating']=imdb_rating[0:100]
IMDB['Year']=yearofrelease[0:100]
```

```python
IMDB
```

```python
IMDB.to_csv('IMDB_Rating.csv',index=False)
```

# 3. Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.  

```python
page2=requests.get('https://www.imdb.com/list/ls009997493/')
```

```python
page2.content
```

```python
soup2=BeautifulSoup(page2.content,'html.parser')
```

```python
Film_Name=[]
Imdb_Rating=[]
Year_Release=[]
```

```python
fname=soup2.find_all('h3',class_='lister-item-header')
fname
```

```python
frating=soup2.find_all('span',class_='ipl-rating-star__rating')
frating[0:100]
```

```python
fyear=soup2.find_all('span',class_='lister-item-year text-muted unbold')
fyear[0:100]
```

```python
for i in frating:
    Imdb_Rating.append(i.get_text().replace("\n",""))
    
Imdb_Rating[0:100]
```

```python
for i in fyear:
    Year_Release.append(i.get_text().replace("\n",""))
    
Year_Release[0:100]
```

```python
import pandas as pd
IMDB_India=pd.DataFrame({})
IMDB_India['Film']=Film_Name[0:100]
IMDB_India['rating']=Imdb_Rating[0:100]
IMDB_India['year']=Year_Release[0:100]
```

```python
IMDB_India
```

```python
IMDB_India.to_csv('IMDB_india.csv',index=False)
```

# 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’ 

```python
page3=requests.get('https://bookpage.com/reviews?book_genre=romance&page=1')
```

```python
page3.content
```

```python
soup3=BeautifulSoup(page3.content,'html.parser')
```

```python
Bookname=[]
Authorname=[]
Genre=[]
```

```python
bname=soup3.find_all('h4',class_='italic')
bname[0:5]
```

```python
aname=soup3.find_all('p',class_='sans bold')
aname[0:5]
```

```python
genre=soup3.find_all('p',class_='genre-links hidden-phone')
genre[0:5]
```

```python
for i in bname:
    Bookname.append(i.get_text().replace("\n",""))
    
Bookname[0:5]
```

```python
for i in aname:
    Authorname.append(i.get_text().replace("\n",""))
    
Authorname[0:5]
```

```python
for i in genre:
    Genre.append(i.get_text().replace("\n",""))
    
Genre[0:5]
```

```python
import pandas as pd
Book=pd.DataFrame({})
Book['name']=Bookname[0:5]
Book['author']=Authorname[0:5]
Book['Genre']=Genre[0:5]
```

```python
Book
```

```python
Book.to_csv('Bookreview.csv',index=False)
```
